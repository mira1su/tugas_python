```python
#operator_aritmatika

#pertambahan
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a + b
print(c)

#pengurangan
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a - b
print(c)

#perkalian
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a * b
print(c)

#pembagian
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a / b
print(c)

#modulus
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a % b
print(c)

#perpangkatan
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a ** b
print(c)

#pembagian bulat
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a // b
print(c)
```

```python
#operator_perbandingan

#samadengan
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a == b
print(c)

#tidaksamadengan
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a != b
print(c)

#lebihbesardari
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a > b
print(c)

#lebihkecildari
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a < b
print(c)

#lebihbesaratausamadengan
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a >= b
print(c)

#lebihkecilatausamadengan
a = int(input("masukkan angka : "))
b = int(input("masukan angka:") )
c = a <= b
print(c)
```

```python
#operator_logika

#truejikakeduapernyataanbenar
bilangan_positif = int(input("masukan angka dari 1-10 : "))
is_positif = bilangan_positif > 0 and bilangan_positif < 10
print(is_positif)


#truejikasalahsatunnyabenar
bilangan_positif = int(input("masukan angka dari 1-10 : "))
is_positif = bilangan_positif > 0 or bilangan_positif < 10
print(is_positif)
```
